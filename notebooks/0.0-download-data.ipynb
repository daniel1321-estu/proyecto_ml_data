```python
from datasets import load_dataset
import os

# Carpeta RAW
os.makedirs("../data/raw/store_sales", exist_ok=True)

# Descarga dataset (HF)
ds = load_dataset("t4tiana/store-sales-time-series-forecasting")

print(ds)
print("Filas TRAIN:", ds["train"].num_rows)
print("Filas TEST:", ds["test"].num_rows)

# Guardar en Parquet (recomendado para 1M+ filas)
ds["train"].to_parquet("../data/raw/store_sales/train.parquet")
ds["test"].to_parquet("../data/raw/store_sales/test.parquet")

print("✅ Guardado en: data/raw/store_sales/")
```
