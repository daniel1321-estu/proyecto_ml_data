# 📈 Análisis Interactivo de Ventas

Este notebook complementa el `dashboard.py` y permite realizar un análisis exploratorio detallado utilizando las mismas lógicas de filtrado y visualización.

```python
import pandas as pd
import plotly.express as px
import os
from pathlib import Path

# Configuración de rutas
PROCESSED_PATH = "../data/processed/store_sales_hf/sales_processed.parquet"

if os.path.exists(PROCESSED_PATH):
    df = pd.read_parquet(PROCESSED_PATH)
    df['date'] = pd.to_datetime(df['date'])
    print("✅ Datos cargados correctamente.")
else:
    print("❌ No se encontraron datos. Ejecuta el pipeline primero.")
```

## 📊 Métricas Globales

```python
total_sales = df['sales'].sum()
avg_sales = df['sales'].mean()
print(f"Ventas Totales: ${total_sales:,.2f}")
print(f"Venta Promedio por Registro: ${avg_sales:,.2f}")
```

## 📈 Tendencia Temporal

```python
daily_sales = df.groupby('date')['sales'].sum().reset_index()
fig = px.line(daily_sales, x='date', y='sales', title='Evolución Diaria de Ventas')
fig.show()
```

## 🏆 Top Categorías

```python
family_sales = df.groupby('family')['sales'].sum().sort_values(ascending=False).head(15).reset_index()
fig = px.bar(family_sales, x='sales', y='family', orientation='h', title='Top 15 Categorías', color='sales')
fig.show()
```
